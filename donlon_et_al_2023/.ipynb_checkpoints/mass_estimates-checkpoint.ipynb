```python
#helper function for reading in data
def get_data_from_fits_pdm(filename, len_segment=600, scan_shift=15):
    #len_segment: number of measurements to put in one segment
    #scan_shift: number of measurements to shift between scans

    #load in all the data and build the dictionaries
    data = {}
    times = []
    fluxes = []
    
    id_tracker = 0
    #should do splits as a uniform amount of time, not n splits per data set
    with open(filename) as f:

        #remove header
        for i in range(3):
            f.readline()

        #read in data
        for line in f:
            line = line.split()
            times.append(float(line[1]))
            fluxes.append(float(line[5]))
    
    i = 0
    while i + len_segment < len(times):
        data[i] = {
                   'times' : times[i:i+len_segment],
                   'fluxes' : fluxes[i:i+len_segment]
        }
        i += scan_shift
    
    return data
```

```python
all_sma = []
all_mass = []
all_sig = []
```

```python
req_sigma = 5.0

def analysis(star_id, prom_tol=0.05):
    
    print(f"star: {star_id}")

    #----------------------------------------

    filename = f"/home/donlon/Desktop/data/kepler_data/SPEX/nonBL/table_kplr{star_id}.tailor-made.dat"
    data = get_data_from_fits_pdm(filename, len_segment=600, scan_shift=15)

    with open(f"/home/donlon/Desktop/data/pickle/maps/maps_v1.1/best_periods_{star_id}_ls600_ss15.pickle", 'rb') as p:
        fit_periods = np.array(pickle.load(p))

    x = [np.nanmean(data[key]['times']) for key in data.keys()]
    x = x - min(x)
    
    fit_uncs = np.zeros(np.shape(fit_periods)) + 0.000006

    indx = np.array([False])
    while len(np.where(np.logical_not(indx))[0]) > 0:
        indx = np.logical_not(np.isnan(fit_periods))*\
               (np.abs(np.array(fit_periods) - np.nanmean(fit_periods)) < 3*np.nanstd(fit_periods))
        
        x = x[indx]
        fit_periods = fit_periods[indx]
        fit_uncs = fit_uncs[indx]

    #-------------------------------------------

    frequency = np.linspace(0.001, 0.2, 100000)
    ls0 = LS(x, np.array(fit_periods), center_data=True) #done this way to save LSP for later
    power = ls0.power(frequency)

    fig = plt.figure(figsize=(20, 5))
    plt.plot(frequency, power, c='k', lw=1)
    plt.xlabel('freq (days$^{-1}$)')
    plt.show()
    
    std = np.std(power)
    #print('standard deviation:', std)
    peak_ind = find_peaks(power, height=req_sigma*std)[0]
    prominences = peak_prominences(power, peak_ind)[0]
    
    for i, pi in enumerate(peak_ind):
        #print(f"period: {1/frequency[pi]} days, significance: {power[pi]/std} sigma")
        #print(f"prominence: {prominences[i]}")
        if star_id == '009717032' or prominences[i] > prom_tol:
            all_sig.append(power[pi]/std)

    pgram_per = 1/frequency
    fig = plt.figure(figsize=(20, 5))
    plt.plot(pgram_per, power, c='k', lw=1)
    ax = plt.gca()
    ylim = ax.get_ylim()
    
    for i, pi in enumerate(peak_ind):
        if star_id == '009717032' or prominences[i] > prom_tol:
            
            print()
            print(f'companion {i}:')
            
            plt.plot([pgram_per[pi]]*2, ylim, c='r', ls='--', lw=1)
            
            #run the LSP best model for this period (gets maximum displacement)
            best_params = ls0.model_parameters(frequency[pi])
            da = (best_params[1]**2 + best_params[2]**2)**0.5
            da = da * 3e8 / np.mean(fit_periods)#(frequency[pi]) #into m/s
            print(f"amplitude: {da} m/s")
            
            #calculate sma
            #assume mass of star is 0.65 Msun
            sma = ((1/(frequency[pi]*365.24))**2 * 0.65)**(1/3) #AU
            print(f"semi-major axis: {sma} AU")
            all_sma.append(sma)
            
            #calculate mass of planet
            period = 1/frequency[pi] * 24*3600 #in seconds
            mass = ((0.65*2e30)**2 * period / (2*np.pi*6.76e-11))**(1/3)*da #Ryden & Peterson
            print(f"mass: {mass} kg ({mass / 1.899e27} JM)")
            all_mass.append(mass)
            
            print()

    plt.xlabel('period (days)')
    plt.xscale('log')
    plt.show()
    
    print('=================================')
```

```python
for star_id, prom in zip(star_ids_list, proms_list):
    analysis(star_id, prom_tol=prom)
```
